```python
# Use urllib to get the url's page data to use with BeautifulSoup
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

import urllib.request
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/Global_Food_Security_Index'
page = urllib.request.urlopen(url)
soup = BeautifulSoup(page, 'lxml')
```

```python
# Gather the tables containing the information we want
tables = soup.find_all('table', class_='wikitable')
table_2022 = tables[0]
```

```python

# Place data from table into lists, to be used in our dataframe object
rank = []
country = []
overall_score = []
affordability = []
availability = []
quality_safety = []
sustainability_adaptation = []

for row in table_2022.find_all('tr'):
    cells = row.find_all('td')
    if len(cells) == 7:
        rank.append(cells[0].text.strip())
        country.append(cells[1].text.strip())
        overall_score.append(cells[2].text.strip())
        affordability.append(cells[3].text.strip())
        availability.append(cells[4].text.strip())
        quality_safety.append(cells[5].text.strip())
        sustainability_adaptation.append(cells[6].text.strip())    
```

```python
# Use a dataframe to store our lists and save it to a csv file

import pandas as pd

df = pd.DataFrame({'Rank': rank})
df['Country'] = country
df['Overall Score'] = overall_score
df['Affordability'] = affordability
df['Availability'] = availability
df['Quality and Safety'] = quality_safety
df['Sustainability and Adaptation'] = sustainability_adaptation

df.to_csv('CSC221-webscrape-data.csv')

df
```
